```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#script editörlerde calısmaz grafik gösteriminde kolaylık saglar
```

```python
yasListesi = [10,20,30,30,40,40,50,60,70,75]
```

```python
kiloListesi = [30,50,70,75,80,85,86,87,88,89]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"y") #renk rgb
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişimi")
#plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20) #0 ile 10 arasında 20 veri 
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak,2 kolon olacak,suan 2.grafik çiziliyor
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni Veri İsimi")
figureAxes.set_ylabel("Y ekseni Veri İsimi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1= figure2.add_axes([0.2,0.2,0.6,0.6])
eksen2= figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni ")
eksen1.set_ylabel("Y ekseni ")
eksen1.set_title("Ana Grafik Başlığı")


eksen2.plot(numpyDizisi1,numpyDizisi2,"b")
eksen2.set_xlabel("X ekseni ")
eksen2.set_ylabel("Y ekseni ")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
    
```

```python
yeniFigur = plt.figure(figsize=(3,3))  #dpi 100 yazılabilir(cozunurluk)
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label = "numpyDizisi ** 3")
yeniEksen.legend(loc=4) #1,2,3,4,5,6 denenebilir farklı yerde yazi
```

```python
yeniFigur.savefig("benimfigur.png",dpi=200)
```
